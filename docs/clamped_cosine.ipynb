# Clamped Cosine

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Definition

$$
\begin{aligned}
f(\theta) &= \max(0, \cos(\theta)) \\ 
f(\vec{v}; \vec{u}) &= \max(0, \vec{v} \cdot \vec{u})
\end{aligned}
$$

```python
def clamped_cosine_1(theta):
    return np.maximum(0, np.cos(theta))

def clamped_cosine_2(u, v):
    return np.maximum(0, np.dot(u,v))
```

## Plot

```python
x_rad = np.linspace(-np.pi, np.pi, 2000)
x_deg = np.rad2deg(x_rad)

y = clamped_cosine_1(x_rad)

fig = plt.figure(figsize=[9,9])
plt.title("Clamped Cosine")
plt.plot(x_deg, y, label="Clamped Cosine")

plt.xlabel("$\\theta$")
plt.ylabel("$f(\\theta)$")
plt.grid()
plt.show()
```
