# Phase Functions

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Polar plot of the Rayleigh phase as a function of theta
def PhaseFunction_Rayleigh(theta):
    cos_theta = np.cos(theta)
    return (3.0/(16.0*np.pi))*(1+cos_theta**2)

theta = np.linspace(-np.pi, np.pi, 100)
ph = PhaseFunction_Rayleigh(theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=[9,9])
ax.plot(theta, ph)
ax.set_rmax(0.15)
ax.set_rlabel_position(0)
plt.title('Rayleigh Scattering')
plt.show()
```

```python
#  Polar plot of the Henyey-Greenstein
def PhaseFunction_HG(theta, g):
    n = 1-g**2
    d = 4*np.pi*np.power(1+g**2-2*g*np.cos(theta),1.5)
    return n/d

# TODO: HG PI?
def PhaseFunction_Schlick(theta, g):
    k = 1.55*g-0.55*(g**3)
    n = 1-k**2
    d = 4*np.pi*(1+k*np.cos(theta+np.pi))**2
    return n/d

g_list = [0, 0.3, 0.6]
theta = np.linspace(0, 2*np.pi, 200)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=[9,9])

for g in g_list:
    ph = PhaseFunction_HG(theta, g)
    ax.plot(theta, ph, label='HG, g={0}'.format(g), color='blue')

for g in g_list:
    ph = PhaseFunction_Schlick(theta, g)
    ax.plot(theta, ph, label='Schlick, g={0}'.format(g), color='red')

ax.set_rmax(0.85)
ax.set_rlabel_position(0)
plt.legend()
plt.show()
```
