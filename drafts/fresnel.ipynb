# Fresnel Reflectance

```python
import math
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

%matplotlib inline
```

$$
F_{\text {Schlick }}\left(F_{0}, \theta\right)=F_{0}+\left(1-F_{0}\right)(1-(cos \theta))^{5}
$$

- $\theta$ : the incoming angle
    - the angle between the light vector and the surface normal
    - also referred to as the incident angle or angle of incidence
- the refractive index of the material

```python
def F_Schlick(F0, theta):
    '''  Schlick approximation  Fresnel term
    F0: 
    theta: the angle between h and l
    '''
    return F0+(1-F0)*math.pow(1-math.cos(theta),5)

F0s = {"Copper":0.95,"Aluminum":0.91,"Iron":0.56,"Diamond":0.17,"Glass":0.08,"Water":0.02}
x_deg = np.linspace(0, 90, 1000)
x_rad = np.deg2rad(x_deg)

plt.figure(figsize=[12,9])
for substance, F0 in F0s.items():
    y = [F_Schlick(F0, i) for i in x_rad]
    plt.plot(x_deg, y, label=substance)

plt.title("Schlick approximation")
plt.xlabel("Incident Angle $\\theta$")
plt.ylabel("Fresnel Reflectance")
plt.legend()
plt.grid()
plt.show()
```
